He decidido empezar el EDA viendo las comparativas de los valores de criptomonedas, U$d, Peso Argentino, etc, en comparación al valor de un Bitcoin, siendo mas claro, por ejemplo, cuantos pesos argentinos equivalen hoy a un Bitcoin, como CURIOSIDAD, todavia no empece a ver mis KPIs elegidos y para ver que tan distantes se encuentran los precios entre si, solo curiosidad, la informacion es extraida de la api coingenko al dia 2023-08-23 (actualidad):

```python
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/exchange_rates"
response = requests.get(url)
exchange_BTC_prices = response.json()

# Accede a la clave "rates" en el JSON
rates_data = exchange_BTC_prices["rates"]

# Convierte el diccionario rates_data en un DataFrame
df = pd.DataFrame.from_dict(rates_data, orient="index")

df = df.sort_values(by="value", ascending=False)

# Formatea la columna "value" con separación de miles y decimales
df["value"] = df["value"].apply(lambda x: "{:,.2f}".format(x).replace(",", " ").replace(".", ",").replace(" ", "."))

# Filtra el DataFrame para quedarte con las filas donde "type" sea "crypto"
df = df[(df["type"] == "crypto") | (df['name'] == "US Dollar")| (df['name'] == "Argentine Peso")]


#hago un drop de la columna type, y el simbolo de la crypto que esta como indice, ya que se que son todas cryptos ahora no nocesito esa info

df.drop(columns="type",axis=1,inplace=True)
df = df.reset_index(drop=True)
print(df.describe())
print("------------------------------")
print(df.info())
# Muestra el DataFrame resultante
df
```

```python
df.to_csv("criptos_a_precio_BTC.csv", index=False)
```
