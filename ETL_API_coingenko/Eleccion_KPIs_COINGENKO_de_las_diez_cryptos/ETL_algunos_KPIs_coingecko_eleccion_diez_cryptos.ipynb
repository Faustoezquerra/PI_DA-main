Aqui algunos criterios segun la api de COINGENKO para elegir las 10 cryptos mas rentables, mas seguras, del mercado, basandonos para empezar por las 13 criptomonedas con mejor <strong>capitalizacion de mercado</strong>, como kpi principal de criptomonedas mas establecidas en el mercado, luego compararemos y analizaremos con los otros 2 KPIs, Volatilización en 30 dias, y Volumen de operaciones en 24hs, y cotejaremos luego con otra api como Binance api, y nos quedaremos con 10 criptomonedas estables, seguras y rentables, pero por ahora nos basamos en esta primera Información para un primer EDA con datos de hoy 25/8/23:

```python
import pandas as pd
import requests

url = "https://api.coingecko.com/api/v3/coins/markets"
parametros = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 13,
    "page": 1
}
respuesta = requests.get(url, params=parametros)
top_13_monedas = respuesta.json()

# Crear una lista para almacenar los datos de todas las criptomonedas
filas_datos = []

for moneda in top_13_monedas:
    id_moneda = moneda["id"]
    url = f"https://api.coingecko.com/api/v3/coins/{id_moneda}?\
        localization=false&tickers=false&market_data=true&community_data=false&developer_data=false&sparkline=false"

    respuesta = requests.get(url)
    datos = respuesta.json()

    datos_mercado = datos.get("market_data")

    if datos_mercado:
        datos_filtrados = {
            "id": id_moneda,
            "simbolo": datos.get("symbol"),
            "nombre": datos.get("name"),
            "fecha_genesis": datos.get("genesis_date"),
            "porcentaje_votos_positivos": datos.get("sentiment_votes_up_percentage"),
            "porcentaje_votos_negativos": datos.get("sentiment_votes_down_percentage"),
            "usuarios_lista_seguimiento": datos.get("watchlist_portfolio_users"),
            "ranking_cap_mercado": datos.get("market_cap_rank"),
            "ranking_coingecko": datos.get("coingecko_rank"),
            "puntuacion_coingecko": datos.get("coingecko_score"),
            "puntuacion_desarrollador": datos.get("developer_score"),
            "puntuacion_comunidad": datos.get("community_score"),
            "puntuacion_liquidez": datos.get("liquidity_score"),
            "puntuacion_interes_publico": datos.get("public_interest_score"),
            "precio_actual_usd": datos_mercado["current_price"]["usd"],
            "precio_maximo_historico_usd": datos_mercado["ath"]["usd"],
            "precio_minimo_historico_usd": datos_mercado["atl"]["usd"],
            "capitalizacion_mercado_usd": datos_mercado["market_cap"]["usd"],
            "valoracion_total_diluida_usd": datos_mercado["fully_diluted_valuation"]["usd"],
            "volumen_total_usd": datos_mercado["total_volume"]["usd"],
            "alto_24h_usd": datos_mercado["high_24h"]["usd"],
            "bajo_24h_usd": datos_mercado["low_24h"]["usd"],
            "cambio_precio_24h_usd": datos_mercado["price_change_24h_in_currency"]["usd"],
            "oferta_total": datos_mercado["total_supply"],
            "oferta_maxima": datos_mercado["max_supply"],
            "oferta_circulante": datos_mercado["circulating_supply"]
        }
    else:
        datos_filtrados = {
            "id": id_moneda,
            "simbolo": datos.get("symbol"),
            "nombre": datos.get("name"),
            "fecha_genesis": datos.get("genesis_date"),
            "porcentaje_votos_positivos": datos.get("sentiment_votes_up_percentage"),
            "porcentaje_votos_negativos": datos.get("sentiment_votes_down_percentage"),
            "usuarios_lista_seguimiento": datos.get("watchlist_portfolio_users"),
            "ranking_cap_mercado": datos.get("market_cap_rank"),
            "ranking_coingecko": datos.get("coingecko_rank"),
            "puntuacion_coingecko": datos.get("coingecko_score"),
            "puntuacion_desarrollador": datos.get("developer_score"),
            "puntuacion_comunidad": datos.get("community_score"),
            "puntuacion_liquidez": datos.get("liquidity_score"),
            "puntuacion_interes_publico": datos.get("public_interest_score"),
        }

    filas_datos.append(datos_filtrados)

# Crear el DataFrame utilizando la lista de datos filtrados
df = pd.DataFrame(filas_datos)

# Rellenar valores nulos con "Sin Dato"
df = df.fillna("Sin Dato")

"""aqui haremos parte del ETL:

Para columnas numéricas:

count: número de valores no nulos.
mean: media aritmética.
std: desviación estándar.
min: valor mínimo.
25%: percentil 25.
50%: mediana o percentil 50.
75%: percentil 75.
max: valor máximo.

Para columnas no numéricas (por ejemplo, tipo object o categorial):

count: número de valores no nulos.
unique: número de valores únicos.
top: valor más común.
freq: frecuencia del valor más común."""

print("------------------------------")
print(df.info())
print("------------------------------")
print(df.describe(include="all"))
print("------------------------------")

# Mostrar el DataFrame resultante
df
```

Eleccion de Criptomonedas Resultante:

**BTC (Bitcoin)**
- **Por qué es buena:** Bitcoin es la primera criptomoneda y es ampliamente reconocida como la principal moneda digital descentralizada. Su adopción masiva y la percepción de ser una reserva de valor le han otorgado el apodo de "oro digital".
- **Tecnología subyacente:** Utiliza una cadena de bloques (blockchain) de prueba de trabajo (PoW, Proof of Work) para procesar y verificar transacciones.
- **De qué se trata:** Bitcoin fue creado por una entidad anónima llamada Satoshi Nakamoto en 2008 y es una moneda digital descentralizada sin intermediarios.

**ETH (Ethereum)**
- **Por qué es buena:** Ethereum introdujo el concepto de contratos inteligentes al mundo de las criptomonedas, permitiendo la creación de aplicaciones descentralizadas (DApps, Aplicaciones Descentralizadas).
- **Tecnología subyacente:** También utiliza blockchain, pero a diferencia de Bitcoin, Ethereum está diseñado para soportar aplicaciones descentralizadas y contratos inteligentes.
- **De qué se trata:** Creado por Vitalik Buterin en 2015, Ethereum es una plataforma que permite a los desarrolladores crear y ejecutar contratos inteligentes y DApps.

**XRP (Ripple)**
- **Por qué es buena:** Ripple está enfocado en hacer que las transferencias transfronterizas sean rápidas y baratas.
- **Tecnología subyacente:** No utiliza blockchain en el sentido tradicional. En su lugar, utiliza un protocolo de consenso distribuido.
- **De qué se trata:** Ripple es tanto una plataforma como una moneda, y tiene el objetivo de permitir transferencias internacionales rápidas y de bajo costo.

**MATIC (Polygon)**
- **Por qué es buena:** Polygon busca solucionar los problemas de escalabilidad de Ethereum y mejorar su usabilidad al transformarlo en una blockchain multi-cadena.
- **Tecnología subyacente:** Es una capa de solución de segunda cadena que utiliza una combinación de redes PoS (Proof of Stake, Prueba de Participación) y PoA (Proof of Authority, Prueba de Autoridad).
- **De qué se trata:** Anteriormente conocido como Matic Network, Polygon proporciona herramientas escalables y compatibles con Ethereum para desarrolladores de DApps y empresas.

**SOL (Solana)**
- **Por qué es buena:** Solana ofrece soluciones de alto rendimiento y baja latencia para aplicaciones cripto.
- **Tecnología subyacente:** Solana utiliza una combinación de PoH (Proof of History, Prueba de Historia) y PoS (Proof of Stake, Prueba de Participación) para mejorar la velocidad y eficiencia de las transacciones.
- **De qué se trata:** Solana es una blockchain de alto rendimiento que busca proporcionar la infraestructura para DApps y criptomonedas descentralizadas.

**ADA (Cardano)**
- **Por qué es buena:** Cardano se describe a sí mismo como una blockchain de tercera generación y se centra en la sostenibilidad, interoperabilidad y escalabilidad.
- **Tecnología subyacente:** Utiliza un algoritmo PoS (Proof of Stake, Prueba de Participación) llamado Ouroboros.
- **De qué se trata:** Creado por uno de los co-fundadores de Ethereum, Charles Hoskinson, Cardano es una blockchain y plataforma de contrato inteligente basada en investigación académica.

**DOGE (Dogecoin)**
- **Por qué es buena:** Aunque comenzó como una broma, Dogecoin ha ganado popularidad por su comunidad activa y es utilizado frecuentemente para dar propinas y donaciones.
- **Tecnología subyacente:** Se basa en el código fuente de Litecoin y utiliza PoW (Proof of Work, Prueba de Trabajo).
- **De qué se trata:** Originalmente creado como una parodia de las criptomonedas basado en el meme "Doge", ha crecido en popularidad y ha sido adoptado en varias campañas benéficas y esfuerzos de crowdfunding.

**BNB (Binance Coin)**
- **Por qué es buena:** Originalmente creada como una token de utilidad para el intercambio Binance, se ha expandido en uso y valor.
- **Tecnología subyacente:** Inicialmente se lanzó en Ethereum, pero luego migró a su propia cadena de bloques, Binance Chain.
- **De qué se trata:** BNB fue creada por Binance, uno de los intercambios de criptomonedas más grandes del mundo, y se utiliza principalmente para pagar tarifas dentro de la plataforma

**DOT (Polkadot)**
- **Por qué es buena:** Polkadot se centra en la interoperabilidad, permitiendo que diferentes blockchains transfieran mensajes y valor de manera descentralizada y segura, facilitando la creación de aplicaciones que trabajan en múltiples cadenas a la vez.
- **Tecnología subyacente:** Polkadot utiliza un modelo multi-cadena con una cadena de retransmisión principal y múltiples cadenas paralelas. Esto permite que las transacciones se procesen en paralelo, aumentando la capacidad de procesamiento.
- **De qué se trata:** Creado por el Dr. Gavin Wood, uno de los co-fundadores de Ethereum, Polkadot tiene como objetivo conectar blockchains privados y públicos, consorcios y oráculos, permitiendo que estas diferentes cadenas compartan información y transacciones mediante su protocolo.

**USDT (Tether)**
- **Por qué es buena:** Tether es una de las primeras y más populares stablecoins (criptomonedas estabilizadas), lo que significa que su valor está atado a una moneda fiduciaria, en este caso, el dólar estadounidense. Esto proporciona estabilidad de precios en comparación con otras criptomonedas volátiles.
- **Tecnología subyacente:** Inicialmente se lanzó en la blockchain de Bitcoin a través del protocolo Omni, pero desde entonces ha sido emitida en otras blockchains, incluidas Ethereum, Tron y EOS.
- **De qué se trata:** Tether busca combinar la estabilidad de precio de las monedas fiduciarias con la libertad de las criptomonedas. Cada USDT emitido teóricamente está respaldado por un dólar estadounidense en reserva, aunque este punto ha sido objeto de debate y controversia.


### KPIs Seleccionados:

1. **Volatilidad del Precio (30 días)**
   - Razón: La volatilidad indica el nivel de riesgo asociado con la inversión en una criptomoneda. Una baja volatilidad puede indicar estabilidad, mientras que una alta volatilidad puede presentar oportunidades de inversión a corto plazo.

2. **Volumen de Comercio en 24h**
   - Razón: Un alto volumen de comercio sugiere una fuerte actividad y liquidez para la criptomoneda. Esto es crucial para los inversores que desean comprar o vender grandes cantidades sin afectar significativamente el precio.

3. **Capitalización de Mercado**
   - Razón: Representa el valor total de una criptomoneda y es un indicador de su popularidad y adopción en el mercado. Una alta capitalización puede ser una señal de confianza en esa criptomoneda.

Estos KPIs, junto con otros datos, se visualizan en el dashboard de Power BI, proporcionando una vista integral del mercado y ayudando a tomar decisiones informadas sobre inversiones en criptomonedas.
