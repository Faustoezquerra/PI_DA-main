```python
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/exchange_rates"
response = requests.get(url)
exchange_BTC_prices = response.json()
```

```python
# Accede a la clave "rates" en el JSON
rates_data = exchange_BTC_prices["rates"]

# Convierte el diccionario rates_data en un DataFrame
df = pd.DataFrame.from_dict(rates_data, orient="index")

# Formatea la columna "value" con separación de miles y decimales
df["value"] = df["value"].apply(lambda x: "{:,.2f}".format(x).replace(",", " ").replace(".", ",").replace(" ", "."))

# Filtra el DataFrame para quedarte con las filas donde "type" sea "crypto"
df = df[df["type"] == "crypto"]

# Muestra el DataFrame resultante
df
```

```python
df.to_csv("criptos_a_precio_BTC.csv", index=False)
```
