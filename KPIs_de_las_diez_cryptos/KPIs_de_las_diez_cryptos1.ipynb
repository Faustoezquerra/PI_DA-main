```python
import requests

url = "https://api.coingecko.com/api/v3/coins/markets"
params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1
}
response = requests.get(url, params=params)
top_10_coins = response.json()
```

```python
import pandas as pd
import requests

for coin in top_10_coins:
    coin_id = coin["id"]
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}?localization=false&tickers=false&market_data=true&community_data=false&developer_data=false&sparkline=false"

    response = requests.get(url)
    data = response.json()

    # Crear una lista para almacenar los datos de cada criptomoneda que se agregar√°n al DataFrame
    data_rows = []

    # Filtrar los datos para conservar solo los valores bajo la clave "usd"
    market_data = data.get("market_data")
    
    if market_data:
                filtered_data = {
                    "id": coin_id,
                    "symbol": data.get("symbol"),
                    "name": data.get("name"),
                    "country_origin": data.get("country_origin"),
                    "genesis_date": data.get("genesis_date"),
                    "sentiment_votes_up_percentage": data.get("sentiment_votes_up_percentage"),
                    "sentiment_votes_down_percentage": data.get("sentiment_votes_down_percentage"),
                    "watchlist_portfolio_users": data.get("watchlist_portfolio_users"),
                    "market_cap_rank": data.get("market_cap_rank"),
                    "coingecko_rank": data.get("coingecko_rank"),
                    "coingecko_score": data.get("coingecko_score"),
                    "developer_score": data.get("developer_score"),
                    "community_score": data.get("community_score"),
                    "liquidity_score": data.get("liquidity_score"),
                    "public_interest_score": data.get("public_interest_score"),
                    "current_price_usd": market_data["current_price"]["usd"],
                    "ath_usd": market_data["ath"]["usd"],
                    "atl_usd": market_data["atl"]["usd"],
                    "market_cap_usd": market_data["market_cap"]["usd"],
                    "fully_diluted_valuation_usd": market_data["fully_diluted_valuation"]["usd"],
                    "total_volume_usd": market_data["total_volume"]["usd"],
                    "high_24h_usd": market_data["high_24h"]["usd"],
                    "low_24h_usd": market_data["low_24h"]["usd"],
                    "price_change_24h_usd": market_data["price_change_24h_in_currency"]["usd"],
                    "total_supply": market_data["total_supply"],
                    "max_supply": market_data["max_supply"],
                    "circulating_supply": market_data["circulating_supply"]
                }
    else: 
             filtered_data = {
                    "id": coin_id,
                    "symbol": data.get("symbol"),
                    "name": data.get("name"),
                    "country_origin": data.get("country_origin"),
                    "genesis_date": data.get("genesis_date"),
                    "sentiment_votes_up_percentage": data.get("sentiment_votes_up_percentage"),
                    "sentiment_votes_down_percentage": data.get("sentiment_votes_down_percentage"),
                    "watchlist_portfolio_users": data.get("watchlist_portfolio_users"),
                    "market_cap_rank": data.get("market_cap_rank"),
                    "coingecko_rank": data.get("coingecko_rank"),
                    "coingecko_score": data.get("coingecko_score"),
                    "developer_score": data.get("developer_score"),
                    "community_score": data.get("community_score"),
                    "liquidity_score": data.get("liquidity_score"),
                    "public_interest_score": data.get("public_interest_score"),
             }

    data_rows.append(filtered_data)

    # Crear el DataFrame utilizando la lista de datos filtrados
    df = pd.DataFrame(data_rows)

    df = df.fillna("Sin Dato")

    df.to_csv(f"{coin_id}_data.csv", index=False)
```
